```python
import pandas as pd
```

```python
xls1 = pd.ExcelFile('irt_stats_interim_2022_23_20220428.xlsx')
irt_valid = pd.read_excel(xls1, 'irt_data')

query = pd.read_csv('_select_distinct_i_item_id_unique_id_wer_dim_t_subject_t_item_gr_202205031124.csv')
```

```python
irt_valid.head()
```

```python
irt_query_results = query[query['item_id'].isin(irt_valid['ItemID'])]
irt_query_results.head()
```

```python
#We notice that there are still duplicate values - tied to wer_dimension columns
irt_query_results['item_id'].value_counts() 
```

```python
#Resetting index
irt_query_results = irt_query_results.reset_index(drop = True)
irt_query_results
```

```python
#Converting unique id's to string values so we can filter by unique id instead of item_id
irt_query_results['unique_id'] = irt_query_results['unique_id'].astype(str)
```

```python
df = irt_query_results[irt_query_results['unique_id'].isin(irt_valid['UniqueID'])]
```

```python
#After checking we see that there are still duplicate values
df['item_id'].value_counts()
```

```python
#This value is an anomaly since it is a duplicate stemming from subject being NaN 
df[df['item_id'] == 94791]
```

```python
#Dropping that row - will be removed from the item_stats table and resetting the index one last time
df = df.drop(194)
df = df.reset_index(drop = True)
```

```python
#Converting Subject to all uppercase for comparison
df['subject'] = df['subject'].str.upper()
```

```python
#Checking data types to match
df.info()
```

```python
irt_valid.info()
```

```python
#Changing data types so they match 
df['item_id'] = df['item_id'].astype(int)
df['ft_year'] = df['ft_year'].astype(int)
df['max_rubric'] = df['max_rubric'].astype(int)
```

```python
#Dropping most recent and OtherFT columns
df = df.drop(columns = 'most_recent')
irt_valid = irt_valid.drop(columns = 'OtherFT')
```

```python
#This loop checks if the values are identical in each dataframe

d = list(df.columns) #list of columns from dataframe generated from item_stats table
v = list(irt_valid) #list of columns from irt_stats file in sharepoint

for n in range(0,len(d)): 
    for i in range(0,len(df)):
        if df[d[n]][i] == irt_valid[v[n]][i]: #This portion checks if each value is identical
            pass
        elif df[d[n]][i] != df[d[n]][i] and irt_valid[v[n]][i] != irt_valid[v[n]][i]: #This portion is error handling for NaN values
            pass
        else:
            print('There is a discrepancy at row' + ' ' + str(i) + ': column' + str(d[n]) ) #If there's a discrepancy, there will be a message
```

```python
df.to_csv('IRT Validated File.csv')
```
